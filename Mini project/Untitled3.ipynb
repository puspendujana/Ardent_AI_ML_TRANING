```python
"""
ğŸ§® Python Calculator
Features: +, -, *, /, %, mean, median, mode, average
"""

from statistics import mean, median, mode, multimode


def get_number(prompt):
    """Prompt user and cast input to float."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("  âš   Please enter a valid number.")


def get_number_list():
    """Prompt user for a comma-separated list of numbers."""
    while True:
        raw = input("  Enter numbers separated by commas (e.g. 3, 5, 7, 5): ")
        try:
            numbers = [float(x.strip()) for x in raw.split(",") if x.strip()]
            if not numbers:
                raise ValueError
            return numbers
        except ValueError:
            print("  âš   Please enter valid comma-separated numbers.")


def basic_calculator():
    """Handle +, -, *, /, % operations."""
    a = get_number("  Enter first number : ")
    print("  Operations: +  -  *  /  %")
    op = input("  Choose operation  : ").strip()
    b = get_number("  Enter second number: ")

    ops = {
        "+": lambda x, y: x + y,
        "-": lambda x, y: x - y,
        "*": lambda x, y: x * y,
        "/": lambda x, y: x / y if y != 0 else None,
        "%": lambda x, y: x % y if y != 0 else None,
    }

    if op not in ops:
        print("  âŒ Unknown operator.")
        return

    result = ops[op](a, b)
    if result is None:
        print("  âŒ Division/modulo by zero is undefined.")
    else:
        # Format: remove unnecessary decimal if whole number
        display = int(result) if result == int(result) else result
        print(f"\n  âœ… {a} {op} {b} = {display}")


def stats_calculator():
    """Handle mean, median, mode, average for a list of numbers."""
    print("\n  Statistical Operations")
    print("  â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    print("  1 â†’ Mean")
    print("  2 â†’ Median")
    print("  3 â†’ Mode")
    print("  4 â†’ Average  (same as mean)")
    print("  5 â†’ All of the above")

    choice = input("\n  Choose option (1-5): ").strip()

    if choice not in {"1", "2", "3", "4", "5"}:
        print("  âŒ Invalid choice.")
        return

    numbers = get_number_list()

    if choice in {"1", "4", "5"}:
        result = mean(numbers)
        label = "Mean / Average" if choice == "5" else ("Average" if choice == "4" else "Mean")
        print(f"\n  ğŸ“Š {label}  : {result:.4f}")

    if choice in {"2", "5"}:
        result = median(numbers)
        print(f"  ğŸ“Š Median  : {result:.4f}")

    if choice in {"3", "5"}:
        modes = multimode(numbers)
        display = ", ".join(str(int(m) if m == int(m) else m) for m in modes)
        print(f"  ğŸ“Š Mode(s) : {display}")


def main():
    print("\n" + "=" * 40)
    print("     ğŸ§®  Python Calculator")
    print("=" * 40)

    while True:
        print("\n  MENU")
        print("  â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print("  1 â†’ Basic Arithmetic  (+, -, *, /, %)")
        print("  2 â†’ Statistics  (mean, median, mode)")
        print("  0 â†’ Exit")
        print("  â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")

        choice = input("  Your choice: ").strip()

        if choice == "1":
            print()
            basic_calculator()
        elif choice == "2":
            stats_calculator()
        elif choice == "0":
            print("\n  ğŸ‘‹ Goodbye!\n")
            break
        else:
            print("  âš   Please enter 0, 1, or 2.")


if __name__ == "__main__":
    main()
```
